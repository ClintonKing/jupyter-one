```python
import pandas as pd
```

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
merged = pd.merge(props, contribs, on='calaccess_committee_id')
```

```python
merged.groupby(['prop_name']).amount.sum().reset_index().sort_values('amount', ascending=False)
```

```python
merged.groupby(['prop_name', 'committee_position']).amount.sum().reset_index()
```

```python
merged.amount.sum()
```

 ~27% of all money was directed either in favor or opposition to prop61

```python
all_donations = merged.contributor_state.value_counts().reset_index()
all_donations
```

```python
donations_total = all_donations.contributor_state.sum()
donations_outside = merged[merged.contributor_state != 'CA'].contributor_state.value_counts().sum()
donations_ca = merged[merged.contributor_state == 'CA'].contributor_state.value_counts().sum()
donations_total
```

```python
donations_outside
```

~4% of all donations were made by parties outside of CA
While the most individual outside donations came from NY, the state with the largest sum of outside money is NJ

```python
cash_by_state = merged.groupby('contributor_state').amount.sum().reset_index().sort_values('amount', ascending=False)
cash_by_state
```

```python
outside_contribs = merged[merged.contributor_state != 'CA']
outside_contribs.groupby(['prop_name', 'committee_position']).amount.sum().reset_index().sort_values('amount', ascending=False)
```

^ Here we see what is almost certainly business influences from out of state.
Largest sums of out-of-state money were directed at opposing state regulation of drug prices, and increase of tax on cigarettes.
Around 86% ($93,714,347) of the total cash opposing prop61 came from out of state donations.

```python
outside_prop61 = outside_contribs[outside_contribs.prop_name == 'PROPOSITION 061- STATE PRESCRIPTION DRUG PURCHASES. PRICING STANDARDS. INITIATIVE STATUTE.']
outside_prop61.groupby(['contributor_state', 'committee_position']).amount.sum().reset_index().sort_values('amount', ascending=False)
```

Most of out-of-state money opposing prop61 came from NJ. 
This is ~40% of out-of-state spending against the prop, and ~35% of all money spent in oppostion to the prop.
No surprise. NJ is home to serveral of the world's largest pharma companies.

```python
nj_no_prop61 = outside_prop61[(outside_prop61.committee_position == 'OPPOSE') & (outside_prop61.contributor_state == 'NJ')]
nj_no_prop61
```
